```python
df = df.drop_duplicates()

print("New Shape:", df.shape)
```

```python
avg_grade = df["G3"].mean()

print("Average Final Grade:", round(avg_grade,2))
```

```python
high_scorers = (df["G3"] > 15).sum()

print("Students Scoring Above 15:", high_scorers)
```

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation:", round(correlation,3))
```

```python
gender_avg = df.groupby("sex")["G3"].mean()

print(gender_avg)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10)

plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.regplot(data=df, x="studytime", y="G3")

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

```python
plt.figure(figsize=(6,4))

gender_avg.plot(kind="bar")

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")

plt.show()
```

# Findings

1. The average final grade (G3) of students is 10.42.

2. A total of 40 students scored above 15, indicating a small group of high-performing students.

3. The correlation between study time and final grade is 0.098, showing a very weak positive relationship between study habits and academic performance.

4. Male students achieved an average grade of 10.91, while female students achieved an average grade of 9.97.

5. The visualizations helped identify grade distribution patterns and compare student performance across different factors.
